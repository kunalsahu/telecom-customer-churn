```python
# Data Processing
import pandas as pd
import numpy as np

# Import necessary modules
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from scipy.stats import randint
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import roc_auc_score
```

Step 1: Data Understanding, Preparation, and Pre-Processing

```python
url = 'https://raw.githubusercontent.com/vikashkmd/DSMLNotebooks/main/loansdata.csv'
data = pd.read_csv(url)
print(data.shape)
data.head()
```

```python
data.describe(include='all')
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
# Missing value treatment

from sklearn.impute import SimpleImputer

# Imputer for numerical columns
num_imputer = SimpleImputer(strategy='median')
```

```python
num_cols_to_impute = ['log.annual.inc', 'days.with.cr.line', 'revol.util', 'inq.last.6mths', 'delinq.2yrs', 'pub.rec']

# Impute the numerical columns
data[num_cols_to_impute] = num_imputer.fit_transform(data[num_cols_to_impute])
```

```python
data.isnull().sum()
```

```python
# Outlier detection and handling using IQR method
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
```

```python
outlier_criteria = ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR)))
```

```python
outlier_columns = ['log.annual.inc', 'dti', 'revol.bal', 'revol.util']
```

```python
outlier_counts = outlier_criteria[outlier_columns].sum()
outlier_counts
```

```python
# Clipping outliers to their 90th percentile value on the higher end
outlier_columns = ['log.annual.inc', 'revol.bal']
percentile_90 = data[outlier_columns].quantile(0.90)

# Clipping the values
for col in outlier_columns:
    data[col] = data[col].clip(upper=percentile_90[col])

# Checking the max values after clipping to ensure clipping was done correctly
max_values_after_clipping = data[outlier_columns].max()
max_values_after_clipping
```

```python
data.head()
```

```python
data_encoded = pd.get_dummies(data, columns=['purpose'], drop_first=True, prefix='Pur')
data_encoded.head()
```

Step 2: Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation matrix
correlation_matrix = data_encoded.corr()

plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Step 3: Feature Engineering and Variable Transformation

```python
from sklearn.preprocessing import MinMaxScaler

# Creating 'Income to Installment Ratio'
# Converting 'log.annual.inc' back to 'annual income' for this calculation
data_encoded['annual_inc'] = np.exp(data_encoded['log.annual.inc'])
data_encoded['income_to_installment_ratio'] = data_encoded['annual_inc'] / (data_encoded['installment'] * 12)

data_encoded.head()
```

```python
# Variable transformation

scaler = MinMaxScaler()

numerical_cols = data_encoded.select_dtypes(include=['float64', 'int64']).columns

data_encoded[numerical_cols] = scaler.fit_transform(data_encoded[numerical_cols])
```

```python
data_encoded.head()
```

Step 4: Model Selection, Model Building, and Prediction

```python
target_column = ['not.fully.paid']
predictors = list(set(list(data_encoded.columns))-set(target_column))
```

```python
X = data_encoded[predictors].values
y = data_encoded[target_column].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=100)
print(X_train.shape); print(X_test.shape)
```

```python
rf = RandomForestClassifier(n_estimators = 50,random_state = 10)
rf.fit(X_train, y_train)
```

```python
predict_train = rf.predict(X_train)
predict_test = rf.predict(X_test)
```

```python
print(confusion_matrix(y_train,predict_train))
print(classification_report(y_train,predict_train))
```

```python
print(confusion_matrix(y_test,predict_test))
print(classification_report(y_test,predict_test))
```

```python
model = RandomForestClassifier(max_depth = 6, n_estimators = 82, random_state = 100, class_weight='balanced')
```

```python
model.fit(X_train, y_train)
```

```python
predict_train = model.predict(X_train)
predict_test = model.predict(X_test)
```

```python
print(confusion_matrix(y_train,predict_train))
print(classification_report(y_train,predict_train))
```

```python
print(confusion_matrix(y_test,predict_test))
print(classification_report(y_test,predict_test))
```
